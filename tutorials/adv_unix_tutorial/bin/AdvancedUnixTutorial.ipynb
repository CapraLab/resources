# 1 | the classics

<br>

These are some basic commands that come in really handy, especially once we know how to combine them (see the next section). But, before we get there-- a quick side note:

````bash
file='../data/E073_18_core_K27ac_mnemonics.bed'
gwas='../data/gwasCatalog_2018-04-01_hg38_n-200.bed'
hg19='../data/Homo_sapiens.GRCh37.75.gtf.gz'
````

This is how we create variables in bash. These are most useful when scripting in bash, but you can create them on the command line as well. To make the following commands simpler, we'll access the filenames using `$file`, `$gwas`, or `$hg19`.

### 1.1 | printing file contents

<br>

Sometimes we want to view the contents of a file or print them to the screen without opening the file in a text editor. These are commands that let you do that.

````bash
cat $file  # print the entire file
````

````bash
head $file  # print first 10 lines of file
````

````bash
tail $file  # print last 10 lines of file
````

Both `head` and `tail` can be modified to print different numbers of lines (default 10) using the `-n` option.

<br>

To page through a file instead, try using:
```
less $file
```

<br>

When debugging pesky BED files, you can use:
```
less -U $file
```

to render tab characters as `^I`. This is really helpful if you're not sure whether your file is space or tab delimited.

### 1.2 | counting lines, cutting columns

<br>

To count the number of lines in a file:

````bash
wc -l $file
````

If you want to use the resulting line count in a later calculation, you can supress the filename print by using `<` to direct the file as input into the command:

````bash
wc -l < $file
````

Sometimes you only want *some* of the columns in a given file. For example, maybe we only want a list of potential chromatin states from our ChromHMM file (column 4):

````bash
cut -f 4 $file  # cut columns 4 from file
````

If we want a range of columns, say the genomic location in a BED file, we can use the following notation:

````bash
cut -f 1-3 $file  # cut columns 1-3 from file
````

### 1.3 | sorting & uniqueness

<br>

We can sort files using the built in `sort` function. All of the options make this pretty powerful, so it's good to get familiar with a few of them. If you have a sorted file, you can find unique elements in it using:

```
uniq $file  # find uniq lines in file
``` 

````bash
sort $file  # sort the file
````

````bash
sort -o '../data/file_sort.out' $file  # write the sorted file to a new file
````

````bash
sort -r $file  # sort the file in reverse order
````

````bash
sort -n $file  # numeric sort the file 
````

````bash
sort -u $file  # sort the file and remove duplicates
````

````bash
sort -k 2 $file  # sort the file by the 2nd column
````

For example, to get a properly sorted BED file we need to sort by the chromosome and start location, and we need to make sure the start coordinates are sorted numerically (not as strings).

````bash
sort -k1,1 -k2,2n $file  # this will give us what we want!
````

You can combine any/all of these options (as above) to create custom sorting commands. This won't change the orginial file, and you can pipe the output to other programs as desired... for more on that, see the next section!

# 2 | combining commands with pipes

<br>

The above commands (and many others) are useful alone, but they are extremely powerful when linked together with pipes (`|`). A pipe takes the standard output of the previous command and uses it as the input to the next command.

````bash
cut -f4 $file | sort -un  # cut the 4th column of the file, sort unique lines numerically
````

As with all kinds of programming, there are many ways to solve any given problem. For example, the command below will sort a BED file by chromosome, then by numeric start, and find uniq coordinates.

````bash
sort -k1,1 -k2,2n $file | cut -f1-3 | uniq
````

Another way to do this might be:

```
cut -f1-3 $file | sort -k1,1 -k2,2n -u 
```

````bash
cut -f1 $file | sort -u | wc -l  # count the number of unique entries in a column
````

Remember these basics because we'll keep using commands from Section 1 and pipes from Section 2 as we move forward!

# 3 | searching for strings with grep

<br>

Global regular expression print... or `grep` is a cool utility that lets us search for lines in a file that match a given regular expression. More on what that means in a minute, but you can think of a regular string as a basic regular expression. So we can use `grep` to search for lines in a file that match a given string.

````bash
grep rs139770682 $gwas  # select lines with a match to this rsID
````

We can get a little fancier with some extra `grep` options:

````bash
grep -i RS139770682 $gwas  # ignore case
````

````bash
grep -v rs139770682 $gwas  # select lines that don't match
````

````bash
grep -c European $gwas  # count matching lines
````

````bash
grep European $gwas | wc -l  # see? same thing!
````

You can also match a number of patterns or regular expressions from a file using the `-f` option.

````bash
echo -e 'Breast cancer\ngut microbiota\nheight' > ../data/test_patterns
grep -f ../data/test_patterns $gwas  # take patterns/regex from a file
````

As with command line utilities in general, you can combine options for increasingly specifc behavior:

````bash
grep -cf ../data/test_patterns $gwas   # count, read patterns from file
````

````bash
grep -cif ../data/test_patterns $gwas  # count, ignore case, read patterns from file
````

To interpret the pattern as an 'extended regular expression' we can use the `-E` option, or call `egrep`. Same thing. This allows us to use more complex regular expressions as patterns.

<br>
For example, we might want to pull out GWAS Catalog associations that are only in Europeans.

````bash
grep -E '[^NR,]\sEuropean(;European)*?\s' $gwas  # interpret pattern as extended regex
````

Confused by that "pattern"? Don't worry-- that's normal. Let's take a quick dive into the world of regular expressions.

# 5 | sed & awk — 'languages' of their own...

<br>

Commands using `sed` and `awk` are powerful-- both because the tools are themselves are powerful **and** because they can be combined with other utilities like `sort` and can use regular expressions.

<br>

### 5.1 | sed

<br>

`sed` stands for 'stream editor' is used to modify 'streams' of text-- often as a find-and-replace tool. For example, we can take that GWAS Catalog file and easily replace the spaces between the first and second columns with a tab character:

````bash
sed 's/ \+/\t/' $gwas  # convert spaces to tabs
````

We could do this in place with the `-i` option, and by specifying a file extension `sed` will save a backup copy of the original file. This is especially useful if you're not 100% sure your regex is going to work the first time.

````bash
sed -i.bak 's/ \+/\t/' $gwas  # convert spaces to tabs in-place and save backup as .bak
````

There is a nice series of intro tutorials here: https://www.digitalocean.com/community/tutorials/the-basics-of-using-the-sed-stream-editor-to-manipulate-text-in-linux.

<br>

Fundamentally, `sed` performs a set of commands on a file or stream that can be altered/customized with various options. The commands include `p` for 'print', `d` for delete, and `s` for substitution.

<br>

You can perform the commands on all lines or on a specific set of lines, using line numbers (`1`) / ranges (`1,+4` or `15-30`) / special characters (`1~10`).

````bash
sed '1d' $gwas  # print file without first line
````

Adding options to `sed` commands increases control over the output. For example, `sed` prints all lines by default. We can supress this with the `-n` option.

````bash
sed '1p' $gwas
````

````bash
sed -n '1p' $gwas
````

Going back to our GWAS Catalog substituion example:

````bash
sed 's/ \+/\t/' $gwas | head # convert spaces to tabs
````

When using the `sed` substitution commands (e.g. `sed 's/ \+/\t/' $gwas`) we need the `/` delimiters to separate the command (`s`) from the regex (` \+`) from the replacement string (`\t`).

<br> 

We can adjust how the substitutions are performed by adding information after the third delimiter. For example, by default `sed` only makes the substitution on the first match in a given line. We can change it to replace **all** instances with `g`. So, if we want to change all ';' delimiters to ',' we could try:

````bash
sed 's/;/,/g' $gwas | head
````

You can ask it to change the 3rd (or nth) instance by swapping the `g` for 3 (or n). Or ask it to ignore case with `i` and print lines that had a substitution with `p`.

````bash
sed 's/European/AnotherEuropean/2p' $gwas | head -20
````

### 5.2 | awk

<br>

`awk` is actually basic scripting language in itself... maybe that's why it seems so complicated? In any case, we'll try to break it down here so you have the building blocks to create your own commands and understand the useful ones that people post on Stack Overflow.

Much like `sed`, `awk` runs a 'script' to do some sort of processing on a file, and there are a number of options you can specify to adjust how that script runs.

````bash
awk '{print}' $gwas  # this is a basic awk script!
````

`awk` scripts are defined by the `'{}'` syntax. We can ask it to print things with the `print` command, this is `$0` by default (which is simply the entire line). We can print specific columns using `$n` where `n` is the column number (1-indexed).

<br>

If we want to use a different delimiter (field separator) then we can use the `-F` option.

````bash
awk -F'\t' '{print $7}' $gwas  # use only tabs, not tabs AND spaces
````

If you want to run multiple commands in a single `awk` script, use a ';' to separate them.

````bash
awk '{$1=="chr13"; print}' $gwas  # print lines with $1 == chr13 only
````

````bash
awk '$1 ~ /chr13/' $gwas  # an alternate approach
````

The above is useful because you can match with regular expressions.

<br>

You can also add text to each line or rearrange columns in a file:

````bash
awk '$1 ~ /chr13/ {print $1,"with start (", $2, ") and end (", $3, ")"}' $gwas  # adding text to printed output
````

````bash
awk -v OFS='\t' '$1 ~ /chr13/ {print $1, $2, $3, $3-$2}' $gwas  # prints region and length, tab-separated
````

````bash
awk 'BEGIN{OFS="\t"} $1 ~ /chr13/ {print $1, $2, $3, $3-$2}' $gwas  # alternatively 
````

Note that `!~` would mean 'does not equal'. But, whoa. We've started to get a bit crazy here. Let's break this down.

<br>

`awk` still has that structure we talked about earlier, we've just expanded it a bit.

```
awk '{script}'
```

expands to

```
awk 'BEGIN{command} /pattern/ {command; command} END{command}'
```
<br>
The things in the BEGIN brackets are run once-- at the beginning of the script. Same with the things in the END brackets. These are good for things like print headers/footers or the outcome of some calculation performed in the script.

````bash
awk 'BEGIN{print "number of bp in this BED file:"} {sum += $3 - $2} END{print sum}' $gwas
````

See? We used the BEGIN block to perform some action at the start. We used our script to sum up the number of bp from each region of the BED input. Then we used our END block to print that sum.

<br>

The OFS thing we did earlier? That's another built in `awk` option. There are a set of useful internal variables that we can set in the BEGIN block, or pass to the script as a variable (`-v OFS='\t'`).
<br><br>

| variable | meaning |
| --- | --- |
|FILENAME | current input file|
|FNR | current record number (if you have multiple input files, this numbers them)|
|FS | field separator for input data (whitespace by default; not just tabs)|
|OFS |output field separator (the FS for the printed output)|
|NR| numer of the current record (i.e. the line)|
|RS| the record separator (newline by default, hence above)|
|ORS| the output record separator (RS for printed output)|

<br>

You can set any of these in an `awk` command, and use any of the potential blocks to create your command. If you want to set the internal variables outside of the '', then you'll need to use the `-v` option to pass it in as a variable. Additionally, if you want a command line variable to be accessible to your `awk` script, you'll need to pass that in as well.

````bash
match_this="chr13"
awk -v var="$match_this" '$1 ~ var {print}' $gwas  # note: we removed the '//' around the pattern
````

You can use other kinds of conditional statements as well: not equal (`!~`), greater than, less than. You can also link multiple together using `&&` or `||`.

````bash
awk '$1 ~ /chr13/ || $4 ~ /4E-9/ {print $1, $2, $3, $4}' $gwas  # on chr 13 OR with p = 4E-9 
````

````bash
awk -v FS='\t' '$1 ~ /chr13/ && $6 > 5000' $gwas  # on chromosome 13 with sample size > 5000
````
